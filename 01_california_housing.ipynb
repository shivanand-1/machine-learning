```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
```

```python
data=fetch_california_housing()
df=pd.DataFrame(data.data,columns=data.feature_names)
print(df.head())
```

```python
df.hist(figsize=(12,10),bins=30,edgecolor="black")
```

```python
plt.figure(figsize=(12,8))
df.boxplot(rot=45)
plt.title("boxplt on numarical python",fontsize=14)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
df.boxplot(rot=45)

plt.ylim(0, 100)   # 👈 LIMIT VALUES TO 0–5000

plt.title("Boxplot on Numerical Data", fontsize=14)
plt.show()
```

```python
def outliers(df):
    dictO={}
    for column in df.columns:
        q1=df[column].quantile(0.25)
        q3=df[column].quantile(0.75)
        iqr=q3-q1
        lower_bond=q1-1.5*iqr
        upper_bond=q3+1.5*iqr
        outliers = df[(df[column] < lower_bond) | (df[column] > upper_bond)]
        outliers = df[(df[column] < lower_bond) | (df[column] > upper_bond)]
        dictO[column]=outliers.shape[0]
print(outliers(df))   
```
