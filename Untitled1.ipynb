```python
#class vs objects
class shivu:
    def __init__(self,name):
        self.name=name
s=shivu("shiu")
print(s.name)
```

```python
#init constructor
class student:
    def __init__(self,name,age):
        self.name=name
        self.age=age
s=student("shivanand",9)
print(s.name)
```
