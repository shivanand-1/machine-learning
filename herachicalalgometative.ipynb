```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
                          
```

```python
data=pd.read_csv("diabetes.csv")
```

```python
df=pd.DataFrame(data)
df
```

```python
X=df.iloc[:,:-1]
```

```python
scaler=StandardScaler()
correct=scaler.fit_transform(X)
```

```python
correct
```

```python
fm=pd.DataFrame(correct,columns=df.iloc[:,:-1].columns)
fm
```

```python
pca=PCA(n_components=2)
d=pca.fit_transform(fm)
```

```python
almrve=AgglomerativeClustering(
        n_clusters=2,
        linkage="ward")
```

```python
model2=almrve.fit(d)
```

```python
inp = list(eval(input("enter the scores")))
k = [[i] for i in inp]  # wraps each number in a list
print(k)
```

```python
label=model2.fit_predict(d)
```
