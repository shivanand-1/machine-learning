```python
import pandas as pd

def sd(d):
    print("Training Data:")

    attri = d.columns[:-1]
    class_label = d.columns[-1]

    hypothesis = ["$" for _ in attri]

    for i, row in d.iterrows():
        if row[class_label] == "Yes":
            for index, value in enumerate(row[attri]):
                if hypothesis[index] == "$" or hypothesis[index] == value:
                    hypothesis[index] = value
                else:
                    hypothesis[index] = "$"

    return hypothesis
d = pd.read_csv("jocs")
(sd(d))
```
