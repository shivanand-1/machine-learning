```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Create Dataset
X, y = make_blobs(
    n_samples=300,
    centers=3,
    random_state=42
)

# Elbow Method
wcss = []

for i in range(1,11):

    kmeans = KMeans(
        n_clusters=i,
        init="k-means++"
    )

    kmeans.fit(X)

    wcss.append(kmeans.inertia_)

print(wcss)

# Final Model
kmeans = KMeans(n_clusters=3)

clusters = kmeans.fit_predict(X)

# Plot
plt.scatter(
    X[:,0],
    X[:,1],
    c=clusters
)

plt.show()
```
